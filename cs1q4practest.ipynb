```python
# cs1q4practest
# Student Score Analyzer Program

def compute_average(scores):
    return sum(scores) / len(scores)

def check_pass_fail(avg):
    if avg >= 75:
        return "Passed"
    else:
        return "Failed"

def main():
    print("=== Student Score Analyzer ===")

    name = input("Enter your name: ")

    # make sure number of quizzes is valid
    while True:
        try:
            num_quizzes = int(input("How many quizzes? "))
            if num_quizzes > 0:
                break
            else:
                print("Please enter a number greater than 0.")
        except:
            print("Invalid input. Enter a number.")

    scores = []

    # loop for scores
    for i in range(num_quizzes):
        while True:
            try:
                score = float(input(f"Enter score for quiz {i+1}: "))
                scores.append(score)
                break
            except:
                print("Invalid input. Enter a number.")

    avg = compute_average(scores)
    result = check_pass_fail(avg)

    print("\n--- Result ---")
    print("Student Name:", name)
    print("Average Score:", round(avg, 2))
    print("Result:", result)

main()
```
